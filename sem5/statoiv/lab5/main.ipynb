```python
import warnings

warnings.filterwarnings("ignore")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import root_mean_squared_error, r2_score
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
PATH = "Walmart.csv"
data = pd.read_csv(PATH)
data.head()
```

```python
data.info()
```

```python
TIME_COLUMN = "Date"
VALUE_COLUMN = "Weekly_Sales"
```

```python
data[TIME_COLUMN] = pd.to_datetime(data[TIME_COLUMN], format="%d-%m-%Y")

data = data[data["Store"] == 1]
data.set_index(TIME_COLUMN, inplace=True)
data = data[VALUE_COLUMN]
```

```python
data.head()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))

plt.plot(data.index, data, label=' ', color='#6A0DAD', linewidth=1)

plt.title('  ', fontsize=14, fontweight='bold')
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend(fontsize=10)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.xticks(fontsize=10, rotation=45)
plt.tight_layout()
plt.show()
```

```python
train_len = 0.8

split_index = int(len(data) * train_len)

train = data.iloc[:split_index]
test = data.iloc[split_index:]
```

```python
model = ExponentialSmoothing(train, trend='add', seasonal='add').fit()
```

```python
predictions = model.forecast(len(test))
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))

train_show = pd.concat([train, test.iloc[:1]], ignore_index=False)
plt.plot(train_show.index, train_show, label=' ', color='#6A0DAD', linewidth=1)

plt.plot(test.index, test, label=' ', color='green', linewidth=1)

plt.plot(predictions.index, predictions, label='', color='yellow', linewidth=1)

plt.title('       ', fontsize=14, fontweight='bold')
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend(fontsize=10)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.xticks(fontsize=10, rotation=45)
plt.tight_layout()

plt.show()
```

```python
difference = test - predictions

difference = difference.dropna() 
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))

plt.plot(difference.index, difference, label=' ( - )', color='blue', linewidth=1)

plt.title('     ', fontsize=14, fontweight='bold')
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend(fontsize=10)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.xticks(fontsize=10, rotation=45)
plt.tight_layout()

plt.show()
```

```python
mse = root_mean_squared_error(test, predictions)
print("RMSE: ", mse)
```

```python
r2 = r2_score(test, predictions)
print("R2: ", r2)
```
