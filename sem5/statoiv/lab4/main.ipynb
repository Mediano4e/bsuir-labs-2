```python
import sklearn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
PATH = "weather_classification_data.csv"
dataset = pd.read_csv(PATH)
dataset.head()
```

```python
dataset.info()
```

```python
cols = []
for column in dataset:
    if type(dataset[column][0]) is str:
        dataset[column] = pd.factorize(dataset[column])[0]
        cols.append(column)
```

```python
dataset = dataset[cols]
```

```python
sns.heatmap(
    round(
        abs(dataset.corr()),
        1,
    ),
    annot=True,
)
```

```python
train_input, test_input, train_output, test_output = train_test_split(
    dataset.drop("Weather Type", axis=1),
    dataset["Weather Type"],
    test_size=0.2
)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(train_input, train_output)
predictions = model.predict(test_input)
```

```python
confusion_matrix = sklearn.metrics.confusion_matrix(predictions, test_output)

sns.heatmap(
    confusion_matrix,
    annot=True
)

plt.title('Матрица сходства')
```

```python
DPI = 450
plt.figure(dpi=DPI)
tree.plot_tree(model)
plt.title("Дерево решений")
```
